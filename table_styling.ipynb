```python
# Importando bibliotecas
import pandas as pd

# Obtendo a url dos dados
url = 'https://raw.githubusercontent.com/Matheus-Pombeiro/dataviz_table_styling/refs/heads/main/relatorio_vendas.csv'

# Lendo os dados
df = pd.read_csv(url)
df
```

```python
df['data_pedido'] = pd.to_datetime(df['data_pedido'])
```

```python
df.info()
```

# Missão 1: Time de vendas

Precisamos construir uma visualização para indicar os **10 maiores clientes por total de venda**.

Com isso, a equipe de vendas pode pensar em estratégias para manter esses clientes comprando em nossa loja.

```python
df_clientes_vendas = df.groupby('nome_cliente')[['vendas']].sum().nlargest(10, 'vendas').copy()
df_clientes_vendas = df_clientes_vendas.reset_index()
df_clientes_vendas
```

```python
df_clientes_vendas.columns = ['Clientes', 'Vendas']
df_clientes_vendas['Rank'] = df_clientes_vendas.index + 1
df_clientes_vendas.set_index('Rank', inplace=True)
df_clientes_vendas
```

```python
s = df_clientes_vendas.style
s
```

```python
s.format({'Vendas': 'R$ {:,.2f}'})
s
```

## Aula 1 - desafio

```python
url = 'https://raw.githubusercontent.com/Matheus-Pombeiro/dataviz_table_styling/refs/heads/main/loja_livro_filmes.csv'
df_loja = pd.read_csv(url)
df_loja.head(2)
```

```python
df_loja_paises = df_loja.groupby('pais')[['unidades']].sum().nlargest(9, 'unidades').copy()
df_loja_paises = df_loja_paises.reset_index()
```

```python
df_loja_paises.columns = ['País', 'Unidades Pedidas']
df_loja_paises['Rank'] = df_loja_paises.index + 1
df_loja_paises.set_index('Rank', inplace=True)
df_loja_paises
```

```python
s_loja = df_loja_paises.style
s_loja.format({'Unidades Pedidas': '{} prod'})
s_loja
```
